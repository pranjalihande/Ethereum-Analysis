```python
import ast
import matplotlib.pyplot as plt

def get_graph(filename, title):

    f = open (filename, 'r')
    h = f.readlines()
    x = ast.literal_eval(h[0])
    x.sort()
    date = []
    value = []

    for i in x:
        date.append(i[0])
        value.append(i[1])

    plt.figure(figsize=(20, 10), dpi=80)
    plt.plot(date, value)
    plt.title(title)
    plt.xlabel('Date: Year-Month')
    plt.ylabel('Value')
    plt.xticks(rotation = 90)
    plt.show()
```

```python
get_graph('average_gas_used_per_month.txt', 'Average gas used per Month')
```

```python
get_graph('time_with_gas_used.txt', 'Time with Gas Used')
```

```python
get_graph('time_with_values_used.txt', 'Time with Values')
```

```python
import pandas as pd
import seaborn as sns
```

```python
with open('./top10address_avg_gas_value.txt', 'r') as file:
    avg_data = file.readlines()

with open('./total_average_gas_used.txt', 'r') as file:
    top10_data = file.readlines()
    
avg_gas_price = eval(top10_data[0])[0][1]
df = pd.DataFrame(eval(avg_data[0]), columns=["Contract", "Value"])
df["Average_Gas_Price"] = df["Value"].apply(lambda x: x[1])
df["Value"] = df["Value"].apply(lambda x: x[0])
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='Contract', y='Average_Gas_Price', data=df)
plt.axhline(y=avg_gas_price, color='b', linestyle='--', label='Average Gas Value')
plt.xlabel('Contract Address')
plt.ylabel('Average Gas')
plt.title('Average Gas for most popular contracts')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
